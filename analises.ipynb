```python
import pandas as pd
dados = pd.read_csv('microdados_enem_2020\DADOS\microdados_enem_2020_l.csv', sep=';')
dados
```

# Perguntas de negócio:
* Qual a escola com a maior média de notas?
* Qual o aluno com a maior média de notas e o valor dessa média?
* Qual a média geral?
* Qual o % de Ausentes?
* Qual o número total de Inscritos?
* Qual a média por disciplina?
* Qual a média por Sexo?
* Qual a média por Etnia?


```python
# * Qual a escola com a maior média de notas?
maior_media_por_escola = dados.loc[:, ['CO_MUNICIPIO_ESC', 'MEDIA_GERAL']]
maior_media_por_escola = maior_media_por_escola.loc[maior_media_por_escola['CO_MUNICIPIO_ESC']>0]
maior_media_por_escola.groupby(['CO_MUNICIPIO_ESC']).mean().sort_values(by='MEDIA_GERAL',ascending=False)
```

```python
# * Qual o aluno com a maior média de notas e o valor dessa média?
maior_media_por_aluno = dados.loc[:, ['NU_INSCRICAO', 'MEDIA_GERAL']]
maior_media_por_aluno.sort_values(by='MEDIA_GERAL',ascending=False)
```

```python
# * Qual a média geral?
media_geral = dados['MEDIA_GERAL'].mean()
media_geral
```

```python
#* Qual o % de Ausentes?
qtd_ausente = dados[['TP_PRESENCA']]
qtd_ausente = qtd_ausente.loc[qtd_ausente['TP_PRESENCA']==0]
porc_ausente = (len(qtd_ausente)/len(dados))*100
porc_ausente
```

```python
# Qual o número total de Inscritos?
len(dados)
```

```python
# Qual a média por disciplina?
# Seleção apenas de quem fez a prova
dados[['NU_NOTA_REDACAO','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC'	,'NU_NOTA_MT']].mean()
				

```

```python
# Qual a média por Sexo?
media_por_sexo = dados.loc[:, ['TP_SEXO', 'MEDIA_GERAL']]
media_por_sexo.groupby(['TP_SEXO']).mean()
```

```python
# Qual a média por Etnia?   
media_por_cor_raca = dados.loc[:, ['TP_COR_RACA', 'MEDIA_GERAL']]
media_por_cor_raca.groupby(['TP_COR_RACA']).mean() 
```
